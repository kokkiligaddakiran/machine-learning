```python
import numpy as np
import pandas as pd
import seaborn as sns 
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
credit_data=pd.read_csv("creditcard.csv")
```

```python
credit_data.head(5)
```

```python
feature_data=credit_data.iloc[:,:-2]
```

```python
feature_data
```

```python
feature_data.size
```

```python
target_data=credit_data.iloc[:,29]
```

```python
target_data
```

```python
sns.heatmap(credit_data.corr())
```

```python
from sklearn.linear_model import LinearRegression
```

```python
linear_model=LinearRegression()
```

```python
linear_model
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,y_train,y_test=train_test_split(feature_data,target_data,test_size=0.9,random_state=0)
```

```python
linear_model.fit(X_train,y_train)
```

```python
y_pred=linear_model.predict(X_test)
```

```python
y_pred
```

```python
linear_model.coef_
```

```python
linear_model.intercept_
```

```python
from sklearn.metrics import r2_score
```

```python
r2_score(y_test,y_pred)
```
